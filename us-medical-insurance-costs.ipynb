```python
# Comparisons of insurance costs by region
import csv

with open('insurance.csv') as insurance_file:
    insurance_data = insurance_file.read()
#     print(insurance_data)
    
list_of_charges = []
list_of_regions = []
northeast_costs = []
northwest_costs = []
southeast_costs = []
southwest_costs = []

with open('insurance.csv', newline = '') as insurance_csv:
    insurance_reader = csv.DictReader(insurance_csv)
    for row in insurance_reader:
        list_of_charges.append(row['charges'])
        list_of_regions.append(row['region'])
        combined_list = zip(list_of_charges, list_of_regions)
        list_of_charges_regions = list(combined_list)
          
    
    for region in list_of_charges_regions:
        if region[-1] == 'northeast':
            northeast_costs.append(float(region[0]))
        elif region[-1] == 'northwest':
            northwest_costs.append(float(region[0]))
        elif region[-1] == 'southeast':
            southeast_costs.append(float(region[0]))
        else:
            southwest_costs.append(float(region[0]))
            
def avg_cost_northeast(northeast_costs):
    sum_northeast = sum(northeast_costs)
    avg_cost = sum_northeast / len(northeast_costs)
    return avg_cost

avg_cost_northeast = avg_cost_northeast(northeast_costs)
print("Northeast Average Cost: " + str(avg_cost_northeast))

def avg_cost_northwest(northewest_costs):
    sum_northwest = sum(northwest_costs)
    avg_cost = sum_northwest / len(northwest_costs)
    return avg_cost

avg_cost_northwest = avg_cost_northwest(northwest_costs)
print("Northwest Average Cost: " + str(avg_cost_northwest))

def avg_cost_southeast(southeast_costs):
    sum_southeast = sum(southeast_costs)
    avg_cost = sum_southeast / len(southeast_costs)
    return avg_cost

avg_cost_southeast = avg_cost_southeast(southeast_costs)
print("Southeast Average Cost: " + str(avg_cost_southeast))

def avg_cost_southwest(southwest_costs):
    sum_southwest = sum(southwest_costs)
    avg_cost = sum_southwest / len(southwest_costs)
    return avg_cost

avg_cost_southwest = avg_cost_southwest(southwest_costs)
print("Southwest Average Cost: " + str(avg_cost_southwest))
              
avg_cost_by_region = {"Northeast": avg_cost_northeast, "Northwest": avg_cost_northwest, 
                      "Southeast": avg_cost_southeast, "Southwest": avg_cost_southwest}
regions_most_to_least_list = sorted(avg_cost_by_region.items(), key=lambda x:x[1], reverse=True)
regions_most_to_least = dict(regions_most_to_least_list)
print(regions_most_to_least)
```
